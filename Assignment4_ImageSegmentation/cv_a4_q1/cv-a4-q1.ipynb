**Cityscapes data** contains labeled videos taken from vehicles driven in Germany. This version is a processed subsample created as part of the [Pix2Pix](https://phillipi.github.io/pix2pix/) paper. The dataset has still images from the original videos, and the semantic segmentation labels are shown in images alongside the original image. This is one of the best datasets around for semantic segmentation tasks.

**Content**

This dataset has 2975 training images files and 500 validation image files. Each image file is 256x512 pixels, and each file is a composite with the original photo on the left half of the image, alongside the labeled image (output of semantic segmentation) on the right half.

### 6. Define Dataset

```python
class CityscapeDataset(Dataset):

  def __init__(self, image_dir, label_model):
    self.image_dir = image_dir
    self.image_fns = os.listdir(image_dir)
    self.label_model = label_model

  def __len__(self):
    return len(self.image_fns)
  
  def __getitem__(self, index):
    image_fn = self.image_fns[index]
    image_fp = os.path.join(self.image_dir, image_fn)
    image = Image.open(image_fp).convert("RGB")
    image = np.array(image)
    cityscape, label = self.split_image(image)
    label_class = self.label_model.predict(label.reshape(-1,3)).reshape(256,256)
    label_class = torch.Tensor(label_class).long()
    cityscape = self.transform(cityscape)
    return cityscape, label_class

  def split_image(self, image):
    image = np.array(image)
    cityscape, label = image[:, :256, :], image[:, 256:, :]
    return cityscape, label
  
  def transform(self, image):
    transform_ops = transforms.Compose([
                                        transforms.ToTensor(),
                                        transforms.Normalize(mean=(0.485, 0.56, 0.406), std=(0.229, 0.224, 0.225))
    ])
    return transform_ops(image)
```

### 8. Train the model

```python
batch_size = 16

epochs = 1
lr = 0.01
```

```python
dataset = CityscapeDataset(train_dir, label_model)
data_loader = DataLoader(dataset, batch_size = batch_size)
```

```python
model = UNet(num_classes = num_classes).to(device)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr = lr)
```

```python
step_losses = []
epoch_losses = []

for epoch in tqdm(range(epochs)):
  epoch_loss = 0
  for X,Y in tqdm(data_loader, total=len(data_loader), leave = False):
    X, Y = X.to(device), Y.to(device)
    optimizer.zero_grad()
    Y_pred = model(X)
    loss = criterion(Y_pred, Y)
    loss.backward()
    optimizer.step()
    epoch_loss += loss.item()
    step_losses.append(loss.item())
  epoch_losses.append(epoch_loss/len(data_loader))
```

**Qualitatively**: We can see that the **with skip connections is able to segment the image better than without skip connections**. This may be attributed to useful information being lost while encoding. With skip connections, information is also shared with deeper layers, so it is not entirely lost. Thus segmentation is better with skip connections.

**Quantitatively**: The scores vary a lot, and sometimes the IoU of without skip connections is more than of with skip connections. This is due to the metric being faulty. It considers all classes collectively (which allows a single majority class to dominate). If the IoU is better in that class for without skip connections, then the overall IoU is dominated and other classes can't contribute much.

```python
def modified_iou(predicted, true):
    iou_list = []
    for cls in range(num_classes):
        pred_idx = (predicted == cls)
        true_idx = (true == cls)
        intersection = np.logical_and(pred_idx, true_idx).sum()
        union = np.logical_or(pred_idx, true_idx).sum()
        
        if union == 0:
                iou_list.append(0)
        else : 
            iou_list.append(intersection / union)
    return np.mean(iou_list)

iou_scores = []


for i in range(test_batch_size):
    
#     landscape = inverse_transform(X[i]).permute(1, 2, 0).cpu().detach().numpy()
    label_class = Y[i].cpu().detach().numpy()
    label_class_predicted = Y_pred[i].cpu().detach().numpy()
    
    
    # IOU score
    iou_score = modified_iou(label_class_predicted, label_class)
    iou_scores.append(iou_score)

print("The mean score of all images across all classes is: " , np.mean(iou_scores))
```

The mean score of all images across all classes is

with skip connections: 0.25042664228901484

without skip connections: 0.2020359138914109

IoU measures the overlap between the predicted and ground truth boundinng box. If perfectly predicted, then IoU is 1.

The above metric was faulty as it was not considering individual classes but rather treating them all as one collectively. This didn't really reflect the models preformance on each class. (It might be doing very well on some simple class which dominates the segmentation. Other class which are not simple to segment,their issue will get hidden under this.)

The mIoU is a better metric as it measures IoU for each class separately. We now calculate the performance of model for each class and the mean represents the overall performance.
Just like IoU, the mIoU also is a metric, but better as it computes IoU for each class separately. It is more reliant, as no single class can dominate the metric. We see that with **skip connections clearly performs better than without skip connections** in mIoU.
